```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Fix path to import training scripts
sys.path.insert(0, os.path.abspath('..'))

# Import training script
from training.train_mlp import main as train_mlp_main

print("Training script imported successfully!")
```

## Step 1: Train MLP Model

```python
# Execute the training script
model, metrics, training_time = train_mlp_main()

print(f"\nModel training completed!")
print(f"Training time: {training_time:.2f} seconds")
```

## Step 2: Model Architecture

```python
print("\n" + "="*70)
print("MLP ARCHITECTURE AND CONCEPTS")
print("="*70)

print("""
Architecture:
  Input (8 features)
    |
    v
  Dense(64) + ReLU activation
    |
  BatchNormalization
    |
  Dropout(0.3)
    |
    v
  Dense(32) + ReLU activation
    |
  Dropout(0.2)
    |
    v
  Dense(16) + ReLU activation
    |
    v
  Dense(1) - Output (MPG prediction)

Key Components:

1. BACKPROPAGATION:
   - Forward Pass: Input -> Layers -> Prediction
   - Loss Computation: MSE = 1/n * Σ(y_true - y_pred)²
   - Backward Pass: Gradients computed using chain rule
   - Weight Update: w_new = w_old - lr * dL/dw

2. ACTIVATION FUNCTIONS:
   - ReLU: max(0, x) - Introduces non-linearity
   - Linear (Output): f(x) = x - For regression tasks

3. REGULARIZATION:
   - L2 Regularization: Penalizes large weights
   - Dropout: Randomly deactivates neurons (30%, 20%)
   - Batch Normalization: Normalizes layer inputs

4. OPTIMIZATION:
   - Adam Optimizer: Adaptive learning rates per parameter
   - Learning Rate: 0.001
   - Early Stopping: Monitor validation loss
   - Learning Rate Decay: Reduce LR when loss plateaus
""")
```

## Step 3: Training Results

```python
print("\n" + "="*70)
print("TRAINING RESULTS")
print("="*70)

print(f"\nModel Name: {metrics['model_name']}")
print(f"Training Time: {training_time:.2f} seconds")
print("\nEvaluation Metrics:")
print(f"  RMSE: {metrics['RMSE']:.4f}")
print(f"  MAE: {metrics['MAE']:.4f}")
print(f"  MSE: {metrics['MSE']:.4f}")
print(f"  R² Score: {metrics['R2']:.4f}")
print(f"  Adjusted R²: {metrics.get('Adjusted_R2', 'N/A')}")
print(f"  MAPE: {metrics['MAPE']:.4f}%")
print(f"  Explained Variance: {metrics['Explained_Variance']:.4f}")
print(f"  Median Absolute Error: {metrics['Median_Absolute_Error']:.4f}")
```

## Step 5: Model Interpretation

```python
print("\n" + "="*70)
print("MODEL INTERPRETATION")
print("="*70)

print("""
1. LOSS CONVERGENCE:
   - Training loss and validation loss decrease over epochs
   - Early stopping prevents overfitting
   - Learning rate reduction helps fine-tuning

2. PREDICTION ACCURACY:
   - RMSE indicates average prediction error in MPG
   - MAE is more robust to outliers than RMSE
   - Low RMSE/MAE indicates good model performance

3. RESIDUAL ANALYSIS:
   - Residuals should be randomly distributed around zero
   - Patterns in residuals indicate systematic error
   - Homoscedasticity: constant variance across predictions

4. ERROR DISTRIBUTION:
   - Should be approximately normal
   - Mean near zero indicates unbiased predictions
   - Narrow distribution indicates consistent accuracy

5. MODEL QUALITY (R² Score):
   - R² = 1: Perfect prediction
   - R² = 0: Model performs as well as predicting mean
   - R² < 0: Model worse than baseline
""")

# Interpret R² score
r2_score = metrics['R2']
if r2_score > 0.9:
    interpretation = "Excellent fit - model explains over 90% of variance"
elif r2_score > 0.7:
    interpretation = "Good fit - model explains over 70% of variance"
elif r2_score > 0.5:
    interpretation = "Moderate fit - model explains over 50% of variance"
else:
    interpretation = "Weak fit - model explains less than 50% of variance"

print(f"\nR² Score Interpretation:")
print(f"  R² = {r2_score:.4f}")
print(f"  {interpretation}")
```
